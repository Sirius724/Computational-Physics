```python
import math as mt
import numpy as np

def f(x): #적분할 함수 지정
    fx = mt.sin(mt.sin(x)) #함수
    return fx

def rectangle(X,dx): #평행사변형 적분 계산
    integ = 0
    for i in range(len(X)-1):
        integ += (f(X[i])+f(X[i+1]))/2*dx
    return integ

a,b = 0, 2*np.pi #적분 시작, 끝 지점 지정
dx = (b-a)/100000 # 계산 간격
X = np.arange(a,b,dx)  # X 구간 설정
integral = rectangle(X,dx) # 적분값 계산

print(f"적분값 : {integral:.5f}")
```

```python
Y = [] 
for i in X:
    Y.append(f(i))
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(X,Y, label = "sin(sin(x))")

Y1 = []
for i in X:
    Y1.append(mt.sin(i))

plt.plot(X,Y1,label = "sin(x)")
plt.legend()
```

```python
def E(T):
    Energy = -4*mt.sinh(4./T)/(12+mt.cosh(4./T))
    return Energy

dx = 1/100
X = np.arange(dx,10+dx,dx)

Y=[]
for i in X:
    Y.append(E(i))

import matplotlib.ticker as ticker

plot1 = plt.axes()
plt.plot(X,Y)
plt.ylim(-4.5,0)
plot1.yaxis.set_major_locator(ticker.MultipleLocator(1))
plt.xlim(0,X[-1]) # X[-1] 리스트 마지막값 반환
plt.show()
```
