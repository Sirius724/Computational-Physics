```python
import numpy as np

path = "/content/drive/MyDrive/Colab Notebooks/전산 물리학"

dx = 1/1000 #미분 간격 

dfx = [] # 미분값 리스트 

# f(x) 함수 정의 
def f(x): 
    fx = x**2
    return fx

save = open("%s/diff_2point.txt"%path,"w+")

X = np.arange(0,1,dx) # X 좌표를 생성
Y = np.array(f(X)) # Y 좌표를 생성 - 데이터가 있다면 데이터로 대체 가능

# 2-point formaular에 의해 f(x)를 미분한 후 출력
for i in range(len(X)-1):
    dfx.append((Y[i+1]-Y[i])/dx)
    save.write("%f  %f  %f\n"%(X[i],Y[i],dfx[i]))


save.close()

```

```python
import pandas as pd
text_test = pd.read_csv("%s/diff_2point.txt"%path, sep='  ',header=None,engine="python")
print(text_test)
from matplotlib import pyplot as plt

plt.plot(X,Y)
plt.plot(text_test[0],text_test[2])
```

```python
# 2차 미분 3점 공식

import numpy as np

path = "/content/drive/MyDrive/Colab Notebooks/전산 물리학"

dx = 1/1000 #미분 간격 

d2fx = [] # 2차 미분값 리스트 

# f(x) 함수 정의 
def f(x): 
    fx = x**3
    return fx

save = open("%s/seconddiff_2point.txt"%path,"w+")

X = np.arange(0,1,dx) # X 좌표를 생성
Y = np.array(f(X)) # Y 좌표를 생성 - 데이터가 있다면 데이터로 대체 가능

# 3-point formaular에 의해 f(x)를 2차 미분한 후 출력
for i in range(1,len(X)-1):
  #print(i)
  d2f =((Y[i+1]-Y[i])/dx-(Y[i]-Y[i-1])/dx)/dx
  save.write("%.5f    %.5f    %.5f\n"%(X[i],Y[i],d2f))

save.close()

```

```python
import pandas as pd

text_test = pd.read_csv("%s/seconddiff_2point.txt"%path, sep='    ', header=None ,engine='python')

print(text_test)

from matplotlib import pyplot as plt


plt.plot(X,Y)
plt.plot(text_test[0],text_test[2])
plt.show()
```

```python
# 특수함수 미분, 2차 미분

import numpy as np
import math as mt

dx = 1/1000 #미분 간격 

d2fx # 2차 미분값 리스트 
dfx  # 1차 미분값 리스트
# f(x) 함수 정의 
def f(x): 
    fx = mt.sin(x)
    return fx

save = open("%s/first.txt"%path,"w+")
save2 = open("%s/second.txt"%path,"w+")

X = np.arange(0,4*mt.pi,dx) # X 좌표를 생성 
Y =[]

for i in range(len(X)):
 Y.append(f(X[i]))# Y 좌표를 생성 - 데이터가 있다면 데이터로 대체 가능


# 2-point formaular에 의해 f(x)를 1차 미분한 후 출력
for i in range(len(X)-1):
    df =((Y[i+1]-Y[i])/dx)
    save.write("%.5f    %.5f     %.5f\n"%(X[i],Y[i],df))

# 3-point formaular에 의해 f(x)를 2차 미분한 후 출력
for i in range(1,len(X)-1):
  #print(i)
  d2f =((Y[i+1]-Y[i])/dx-(Y[i]-Y[i-1])/dx)/dx
  save2.write("%.5f    %.5f    %.5f\n"%(X[i],Y[i],d2f))

save.close(); save2.close()

import pandas as pd

text_test = pd.read_csv("%s/first.txt"%path, sep='    ', header=None ,engine='python')
text_test2 = pd.read_csv("%s/second.txt"%path, sep='    ', header=None ,engine='python')

from matplotlib import pyplot as plt

plt.xlim(0,4); plt.ylim(-1,1)
plt.plot(X/mt.pi,Y,"r",label = "Sin(x)")
plt.plot(text_test[0]/mt.pi,text_test[2],"black",label = "First derivative")
plt.plot(text_test2[0]/mt.pi,text_test2[2],"blue", label = "Second derivative")
plt.legend(loc = "upper right")
plt.show()
```
