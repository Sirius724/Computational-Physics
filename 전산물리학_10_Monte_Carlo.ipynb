# Monte Carlo integration method
## Random number generator + integration


$$ A ∝ \frac{N_{in}}{N_{total}} $$

```python
# Find Pi value by monte carlo integration method

def func(x):
    fun = np.sqrt(1-x**2)
    return fun

from numpy import random

try_number = 5000 # 시도 횟수 
x = random.random(size = try_number) # 0에서 1 사이 실수 리스트 생성
y = random.random(size = try_number)

result_list = np.array(func(x))

count = 0
for i in range(try_number):
    #result_list[i]= func(x[i])
    if result_list[i] > y[i]:
        count +=1

print(f"the number of attempts = {try_number}, pi = {4*count/try_number}")


plt.figure(figsize=(8,8))
plt.xlim(0,1)
plt.ylim(0,1)
plot_X = np.linspace(0,1,100)
plt.scatter(x,y,alpha=0.5,c="r",s =2)
plt.plot(plot_X,func(plot_X),linewidth=3)
plt.title("Monte carlo Pi",size = 20)
plt.text(0.1,0.5,f"the number of attempts = {try_number}, pi = {4*count/try_number}",fontsize=10)
plt.show()
```
